**Pandas Series**

A Pandas Series is like a column in a table. It is a 1-D array holding data of any type.

Importing Pandas

```python
import pandas as pd
import numpy as np
```

**Series from lists**

```python
#string

country = [ 'India','Pakistan','USA','Nepal','Srilanka']
pd.Series(country)
```

```python
# Integer
runs = [100,200,300,400,500]
run_series = pd.Series(runs)
run_series
```

```python
# custom index
marks = [10,20,30,40,50]
subjects = ['Englis','Maths','Science','Social','Hindi']
pd.Series(marks,index = subjects)
```

```python
# setting a name
marks = pd.Series(marks,index = subjects,name = 'pankaj ke marks')
marks
```

**Series from dict**

```python
marks = {'Englis':10,'Maths':20,'Science':30,'Social':40,'Hindi':50}
marks_series = pd.Series(marks,name = 'pankaj ke marks')
marks_series
```
